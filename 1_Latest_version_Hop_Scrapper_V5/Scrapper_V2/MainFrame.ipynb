```python
from CallerPy import process_urls
import pandas as pd

# Beispielaufruf
initial_urls = ["https://www.eversports.de/s/poda-studio", "https://www.eversports.de/s/css-aerial-dance-studio"]

# Dictionary, um die Ergebnisse zu speichern
all_data = {}

# Schleife √ºber jede URL in der Liste
for url in initial_urls:
    print(f"Verarbeite URL: {url}")
    data = process_urls([url])  # Achten Sie darauf, eine Liste zu √ºbergeben
    all_data[url] = data

# Zugriff auf die verschiedenen DataFrames f√ºr jede URL
for url, data in all_data.items():
    print(f"\nErgebnisse f√ºr URL: {url}")

    # Pole Studio Daten
    if 'pole_studio_data' in data:
        pole_studio_df = data['pole_studio_data']
        print("Pole Studio Daten:")
        print(pole_studio_df)

    # Workshops Daten
    if 'workshops_data' in data:
        workshops_df = data['workshops_data']
        print("Workshops Daten:")
        print(workshops_df)

    # Workshop Details
    if 'workshop_details' in data:
        workshop_details_df = data['workshop_details']
        print("Workshop Details:")
        print(workshop_details_df)

    # Klassen Daten
    if 'classes_data' in data:
        classes_df = data['classes_data']
        print("Klassen Daten:")
        print(classes_df)

    # Klassen Details
    if 'classes_details' in data:
        classes_details_df = data['classes_details']
        print("Klassen Details:")
        print(classes_details_df)
```
